```python
fig = plt.figure(figsize = (12,8))
ax1 = fig.add_subplot(221)
imshow(img)
ax1.set_title('Shape: '+ str(img.shape));

ax2 = fig.add_subplot(222)
new_image = cv2.flip(img,0) # 0: vertial, 1:horizontal
imshow(new_image)
ax2.set_title('Shape: '+ str(new_image.shape));

ax3 = fig.add_subplot(223)
new_image = cv2.flip(img,1) 
imshow(new_image)
ax3.set_title('Shape: '+ str(new_image.shape));

ax4 = fig.add_subplot(224)
new_image = cv2.flip(img,-1) 
imshow(new_image)
ax4.set_title('Shape: '+ str(new_image.shape));
```
